```python
# Importar libreras necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar estilo general para los grficos
sns.set(style="whitegrid")

# Cargar el archivo directamente desde la ruta del entorno de Google Colab
file_path = '/content/data.csv'
df = pd.read_csv(file_path)



# Crear nueva columna de descuento
df['discount'] = df['price'] - df['offer_price']

# 1. Grfico de barras: Precio promedio por marca (Top 10 marcas)
avg_price_by_brand = df.groupby('brand')['price'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
avg_price_by_brand.plot(kind='bar')
plt.title('Precio Promedio por Marca (Top 10)')
plt.xlabel('Marca')
plt.ylabel('Precio Promedio')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2. Grfico de barras: Descuento promedio por marca (Top 10 marcas)
avg_discount_by_brand = df.groupby('brand')['discount'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
avg_discount_by_brand.plot(kind='bar', color='orange')
plt.title('Descuento Promedio por Marca (Top 10)')
plt.xlabel('Marca')
plt.ylabel('Descuento Promedio')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 3. Grfico de conteo: Distribucin de tallas segn colores predominantes
top_colors = df['color'].value_counts().head(5).index
df_top_colors = df[df['color'].isin(top_colors)]

plt.figure(figsize=(12, 6))
sns.countplot(data=df_top_colors, x='size', hue='color', order=sorted(df['size'].unique()))
plt.title('Distribucin de Tallas por Colores (Top 5 Colores)')
plt.xlabel('Talla')
plt.ylabel('Cantidad')
plt.legend(title='Color')
plt.tight_layout()
plt.show()


```
